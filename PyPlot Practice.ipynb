```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
apple_data = pd.read_csv('AAPL.csv')
apple_data
```

```python
dates = apple_data['Date']
open = apple_data['Open']
high = apple_data['High']
low = apple_data['Low']
close = apple_data['Close']
```

```python
plt.plot(dates,close)
plt.plot(dates,open)
plt.show()
```

```python
plt.plot(dates, high)
plt.plot(dates, low)
plt.show()
```

```python
fig = plt.figure()
plt.subplot(411)
plt.plot(dates,open)
plt.subplot(412)
plt.plot(dates,close)
plt.subplot(413)
plt.plot(dates,high)
plt.subplot(414)
plt.plot(dates,low)
plt.show()
```

```python
figure,axes = plt.subplots(2,2)
axes[0,0].plot(dates,open)
axes[0,1].plot(dates,close)
axes[1,0].plot(dates,high)
axes[1,1].plot(dates,low)
plt.show()
```

```python
#labeling
plt.plot(dates,close)
plt.title('ClosePrices of AAPL')
plt.xlabel('Dates')
plt.ylabel('Prices in USD')
plt.xticks(pd.date_range('2019-05-14','2020-05-13',5).strftime('%Y-%m-%d'))
plt.show()
```

```python
figure,axes = plt.subplots(1,2)
fig.suptitle('AAPL Prices',y=.005)
axes[0].plot(dates,high)
axes[0].set_title('Highs for AAPL')
axes[0].set_ylabel('Price')
axes[0].set_xlabel('Date')
axes[1].set_title('Lows for AAPL')
axes[1].set_ylabel('Price')
axes[1].set_xlabel('Date')
axes[1].plot(dates,low)
plt.tight_layout()
plt.show()
```

```python
plt.plot(dates,high)
plt.plot(dates,low)
plt.title('Daily High and Low of AAPL')
plt.xlabel('Dates')
plt.ylabel('Prices in USD')
plt.xticks(pd.date_range('2019-05-14','2020-05-13',5).strftime('%Y-%m-%d'))
plt.legend(['High','Low'])
plt.show()
```
